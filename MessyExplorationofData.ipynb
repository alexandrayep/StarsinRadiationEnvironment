```python
import numpy as np
import astropy.io.fits as fits
import glob
import matplotlib.pyplot as plt
import astropy

#wavelength, flux
def wf(dat): #whichord,dat. Setup:   w,f=wf(#,dat)
    w=np.array([d[0] for d in dat])
    f=np.array([d[1] for d in dat])
    return w,f
```

```python
for file in sorted(glob.glob("*.fits")):
    hdu=fits.open(file)
    data=hdu[0].data
    head=hdu[0].header
    hdu.close()
    print('\n'+file+':')
    print(head['OBJECT'])
    print(head['WAVMODE'])
    #print(head['GSP_FNAM'])
    for h in head:
        if 'LAMP' in h:
            print(h,head[h])
    plt.figure()
    plt.imshow(data)
    plt.title(file)
```
